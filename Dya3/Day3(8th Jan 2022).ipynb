#### histogram
- represented as bins
- bins represents the data points in the range
- plt.hist(ele,bins=count)

```python
import numpy as np
std1=np.random.randint(18,45,100)
#std1
std2=np.random.randint(15,40,100)
#std2
plt.hist(std1,bins=50)
plt.xlabel('age category')
plt.ylabel("no.of students")
plt.show()
```

```python
plt.hist(std2,bins=10,color='c',orientation='horizontal')
```

```python
std1
```

```python
std2
```

```python
vals=[num for num in range(10,50,5)]
plt.hist(std1,vals,histtype='bar',
         orientation='vertical',color='m')
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
plt.figure(figsize=(8,5))
plt.hist([std1,std2],vals,rwidth=0.7,
         histtype='bar',orientation='vertical',
         color=['m','y'])
plt.title("Multiple histograms")
```

#### pie plot

```python
x=[56,25,30,24] # making the regions
plt.pie(x)
plt.show()
```

```python
#adding labels
fb=['average','good','above average','Excellent']
plt.pie(x,labels=fb) # assigning labels of fb to x
plt.show()
```

```python
#adding colors to regions of pie
plt.pie(x,labels=fb,
        colors=['green','cyan','blue','pink'])
plt.title("Sort by Responses")
plt.show()
```

```python
#adding colors to regions of pie
plt.pie(x,labels=fb,
        colors=['green','cyan','blue','pink'],
       autopct="%.f%%") #assigning % to regions
plt.title("Sort by Responses")
plt.show()
```

```python
#adding shadow to regions of pie
ex=[0,0,0,1]
plt.pie(x,labels=fb,
        colors=['green','cyan','blue','pink'],
       autopct="%.f%%",shadow=True,startangle=180,
       explode=ex) # excluding part
plt.title("Sort by Responses")
plt.show()
```
